#Feature Engineering

```python
Y = data[['Hujan','SinarMatahari','Awan3pm','Suhu3pm']] #ambil kolom yang diperlukan sesuai readme.txt
Y
```

```python
ff = data[['Hujan','SinarMatahari','Awan3pm','Suhu3pm']] #ambil kolom yang diperlukan sesuai readme.txt
ff
```

```python
ff.columns = ['Hujan','SinarMatahari','Awan3pm','Suhu3pm'] #ambil data sesuai nama kolom tercantum
Y = ff.values #ambil nilai dari data
plt.scatter(Y[:,0],Y[:,1],Y[:,2],Y[:,3]) #plotting data sesuai indeks masing masing
plt.show() #tampilkan plotting
```

```python
X = ff.values #ambil nilai data
m = Y.shape[0] #ubah dimensi
n = Y.shape[1] #ubah dimensi
print("No. of data points: ",m,"\nNo. of data features:",n)
K = int(input("Enter the value of K: "))
```

```python
import random as rd #generate

def k_means(K, c):  #fungsi k means untuk clustering
    ED = np.zeros((m, 0))
    out = {} #temporary output untuk iterasi 
    for k in range(K):
        tempDist = np.sqrt(np.sum((X - c[:,k])**2, axis=1)) #rumus k means
        ED = np.c_[ED, tempDist] #rumus euclidians distance
        out[k] = np.zeros((n, 0))
    c_temp = np.argmin(ED, axis=1)
        
    for i in range(m):
        out[c_temp[i]] = np.c_[out[c_temp[i]], X[i]]
    
    for k in range(K):
        out[k] = out[k].T
  
    for k in range(K):
        c[:,k] = np.mean(out[k], axis=0)
    return out, c
```

```python
c = np.zeros((n, 0))
for i in range(K):
    rand = rd.randint(0,m-1)
    c = np.c_[c, X[rand]]
print("Centroids before clustering: \n", c)

Output = {} #kamus setiap cluster, dengan nomor cluster sebagai kunci kamus, dan titik data milik cluster sebagai nilai yang sesuai
iterations = 200
for i in range(iterations):
  Output, c = k_means(K, c)
print("\nCentroids after clustering: \n", c)
```

#Visualisasi Setelah Clustering

```python
color = ['red', 'blue', 'cyan', 'green', 'magenta']
labels = ['cluster 1', 'cluster 2', 'cluster 3', 'cluster 4', 'cluster 5']
for k in range(K):
    plt.scatter(Output[k][:,0], Output[k][:,1], c=color[k], label=labels[k])
plt.scatter(c[0,:], c[1,:], s=100, c='yellow', label='Centroids')
plt.legend()
plt.show()
```
